**1.Importing Libraries**

```python
import pandas as pand #data manip
import matplotlib.pyplot as plt #data visualization
```

**2. Loading Dataset**

```python
dataset = pand.read_csv('sales_data_2.csv')
```

**3.Data loading and initial analysis**

```python
stat = dataset.describe()
```

**4. Data manipulation**

```python
product_rev = dataset.groupby('Product')['Revenue'].sum()
highest_rev = product_rev.idxmax()
avg_qual_day = dataset['Quantity Sold'].mean()

print (product_rev)
print (" ")
print (highest_rev)
print (" ")
print (avg_qual_day)
```

**5.Data visualization**

```python
product_rev.plot(kind = 'bar', color = 'orange')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show
```

**6. Analysis Report**

```python
report = """
Analysis Report

1.  Data Loading and initial Analysis:
    Basic Statistics of the dataset:
    {}
2.  Data Manipulation:
    Total revenue for each prodcut:
    {}
    Product with highest revenue:{}
    Avaerage quantity sold per day: {:.2f}
3.  Data Visualnization
    Bar chart diaplaying revenue for each product is attached
4.  Insights:
  -The product{} generates the highest revenue.
  -On average, {:.2f} units sold per day
""".format (stat,product_rev,highest_rev,avg_qual_day,highest_rev,avg_qual_day)

print(report)
```

**7. Saving the report to a file**

```python
with open ('report.txt', 'w') as f:
    f.write(report)
```
