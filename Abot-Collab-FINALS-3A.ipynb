```python
from automata.fa.nfa import NFA
```

```python
nfa_example = NFA(
    states= {'q0','q1','q2'},
    input_symbols={'0','1'},
    transitions= {
        'q0':{'0': {'q1'}, '1': {'q2'}},
        'q1':{'1':{'q1'} ,'0':{'q2'}},
        'q2': {}
    },
    initial_state= 'q0',
    final_states={'q2'}
)
```

```python
def accepted_ex (input_str):
  return nfa_example.accepts_input(input_str)
```

```python
test = ['0100','00','111100','01110','11110']

for test_str in test:
    if accepted_ex(test_str):
      print(f'String"{test_str}" is accepted.')
    else:
      print(f'String"{test_str}"is not accepted.')
```

**CONVER NFA TO DFA**

```python
from automata.fa.dfa import DFA
```

```python
dfa = DFA(
    states={'q0', 'q1', 'q2', 'q0q1', 'q0q2', 'q1q2'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': 'q0q1', '1': 'q2'},
        'q1': {'0': 'q2', '1': 'q1'},
        'q2': {'0': 'q2', '1': 'q2'},
        'q0q1': {'0': 'q2', '1': 'q1q2'},
        'q0q2': {'0': 'q2', '1': 'q2'},
        'q1q2': {'0': 'q2', '1': 'q0q2'}
    },
    initial_state='q0',
    final_states={'q2', 'q0q2', 'q1q2'}
)
```

```python
def accept (input_str2):
  return dfa.accepts_input(input_str2)
```

```python
test_strings = ['00','1111','01011','111']

for test_str2 in test_strings:
  if accept(test_str):
    print(f'String "{test_str2}" is accepted')
  else:
    print(f'String "{test_str2}" is not accepted')
```
