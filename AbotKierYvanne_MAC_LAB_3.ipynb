#**1.Import libraries**

```python
import pandas as pand
import numpy as nump
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# **2. Generating Random Data**

```python
num_points = 50

x = nump.random.randint(0,100,num_points)
y = 2*x+nump.random.normal(0,10,num_points)
```

#**3. Creating Dataframe and saving to csc**

```python
data = pand.DataFrame({'x':x,'y':y})

data.to_csv('data.csv',index = False)

print(data.head())
```

#**4. Visualizing The Data**

```python
plt.scatter(data['x'],data['y'])
plt.xlabel('indipendent variable')
plt.ylabel('dependent variable')
plt.title('scatter of data')
plt.show()
```

#**5. Preparing Data for Modeling**

```python
x = data[['x']]
y = data[['y']]
```

#**6. Fitting The linear regression model**

```python
model = LinearRegression()
model.fit(x,y)
```

#**Getting model coefficents**

```python
slope = model.coef_[0]

intersept = model.intercept_

print('slope',slope)

print('intersept',intersept)
```

#**8. Visualizing the linear regression line**

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x), color ='green')
plt.xlabel('indipendent variable')
plt.ylabel('dependent variable')
plt.title('scatter of data')
plt.show()
```

#**9. Making Prediction**

```python
new_x = [[nump.random.randint(100,200)]]

predicted_y = model.predict(new_x)

print("value of x:",new_x)
print('Predicted Value of Y:',predicted_y)
```
