#**1. Import Library**

```python
import pandas as pand
import numpy as nump
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

#**2. Creating Synthetic Dataset**

```python
nump.random.seed(0)
x = nump.random.rand(100,3)
y = x[:,0]*150000 + x[:,1]*100000+x[:,2]*50000+50000
```

#**3. Convering to dataframe**

```python
data = pand.DataFrame(x,columns = ['size','bedroom','age'])
data['price'] = y

data.head()
```

#**4. Preparing Data for modeling**

```python
x = data[['size','bedroom','age']]
y = data['price']
```

#**5. Training and split data**


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 42)
```

#**6.Fitting linear regression model**

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

#**7.Making Prediction**

```python
y_predict = model.predict(x_test)
```

#**8. Mean squeared error**

```python
mse = mean_squared_error(y_test,y_predict)

print(mse)
print(y_predict)
```
