#**1.import libraries**

```python
import pandas as pand
import numpy as nump
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# **2. Reading csv**

```python
data = pand.read_csv('Lesson_4_student_grades_dataset.csv')
```

#**3.Visualize data**

```python
plt.scatter(data['Previous_Exam_Scores'],data['Grades'])
plt.xlabel('indipendent variable')
plt.ylabel('dependent variable')
plt.title('scatter of data')
plt.show()
```

#**4.Visualizing Linear Regression**

```python
x = data[['Previous_Exam_Scores']]
y = data[['Grades']]

model = LinearRegression()
model.fit(x,y)

slope = model.coef_[0]
intersept = model.intercept_
print('slope',slope)
print('intersept',intersept)

plt.scatter(x,y)
plt.plot(x,model.predict(x), color ='purple')
plt.xlabel('indipendent variable')
plt.ylabel('dependent variable')
plt.title('scatter of data')
plt.show()
```

#**5. Making Predition**

```python
new_x = [[data['Previous_Exam_Scores'][12]]]

predicted_y = model.predict(new_x)

print("value of x:",new_x)
print('Predicted Value of Y:',predicted_y)
```
