#1. import libraries**

```python
import pandas as pand
import numpy as nump
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

#**2. Getting data from csv**

```python
data = pand.read_csv('Lesson_4_student_grades_dataset.csv')
```

#**3. Creating Dataframe**

```python
df = pand.DataFrame(data)
x = df.drop(['Grades'],axis = 1)
y = df['Grades']

df.head()

x = data[['Hours_of_Study','Attendance_Percentage','Previous_Exam_Scores','Extracurricular_Activities','Travel_Time_to_School']]
y = data['Grades']

data.head()
```

#**4. Training split data**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.60, random_state = 42)
```

#**5.Linear Modeling**

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

#**6. Making Prediction**

```python
new_data = nump.array([[5,90,98,1,50]])
prediction = model.predict(new_data)

grade = (prediction/60)*100

print(grade)
y_predict = model.predict(x_test)
```

#**7. Mean squared error**

```python
mse = mean_squared_error(y_test,y_predict)

print(mse)
print(y_predict)
```
