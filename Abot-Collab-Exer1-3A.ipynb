**2. Import functions**

**3. Example of an NFA and code**

 **Define NFA With Its State**


*  **State:**
*   **Input Symbols:**
*   **Transitions:**
*   **Intial State:**
*   **Final State**


NFA.png






```python
nfa_example = NFA(
    states= {'q0','q1','q2'},
    input_symbols={'0','1'},
    transitions= {
        'q0':{'0': {'q1'}, '1': {'q0'}},
        'q1':{'1':{'q1'} ,'0':{'q2'}},
        'q2': {}
    },
    initial_state= 'q0',
    final_states={'q0','q2'}
)
```

**4. Define functiion to check if the given input string is NFA**

```python
def accepted_ex (input_str):
  return nfa_example.accepts_input(input_str)
```

**5. test string**

```python
test = ['100','00','111100','01110','11110']

for test_str in test:
    if accepted_ex(test_str):
      print(f'String"{test_str}" is accepted.')
    else:
      print(f'String"{test_str}"is not accepted.')
```
